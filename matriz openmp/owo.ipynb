```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Dados de exemplo
threads_mpi = [1, 4, 13, 26]  # Número de threads
temposMPI = [12.408, 5.831, 2.891, 1.931]  # Tempo correspondente
tempos_1s = [12.495,5.717,3.939]  
tempos_1g = [12.509,4.374,2.929]
tempos_1d = [12.503,3.891,2.686]
tempos_13s = [2.583,2.572,2.545]
tempos_13g = [2.559,2.561,2.540]
tempos_13d = [2.533,2.526,2.545]
tempos_26s = [1.982,1.971,1.974]
tempos_26g = [1.982,1.949,1.961]
tempos_26d = [1.974,1.958,2.018]
# Criar o gráfico de linhas
plt.figure(figsize=(8, 5))
plt.plot(threads_mpi, temposMPI, marker='o', linestyle='-', color='b', label='Tempo MPI puro')


# Configurar os rótulos e título
plt.xlabel('Número de Processos')
plt.ylabel('Tempo (segundos)')
plt.title('Gráfico de Linhas: Processos vs Tempo')
plt.grid(True)
plt.legend()

# Exibir o gráfico
plt.show()
```

```python
# Criar o gráfico de linhas
threads = [1,4,8]  # Número de threads
plt.figure(figsize=(8, 5))
plt.plot(threads, tempos_1s, marker='o', linestyle='-', color='b', label='Tempo 1 processo Static')
plt.plot(threads, tempos_1d, marker='o', linestyle='-', color='r', label='Tempo 1 processo Dynamic')
plt.plot(threads, tempos_1g, marker='o', linestyle='-', color='g', label='Tempo 1 processo Guided')
plt.plot(threads, tempos_13s, marker='o', linestyle='-', color='c', label='Tempo 13 processos Static')
plt.plot(threads, tempos_13d, marker='o', linestyle='-', color='m', label='Tempo 13 processos Dynamic')
plt.plot(threads, tempos_13g, marker='o', linestyle='-', color='y', label='Tempo 13 processos Guided')
plt.plot(threads, tempos_26s, marker='o', linestyle='-', color='k', label='Tempo 26 processos Static')
plt.plot(threads, tempos_26d, marker='o', linestyle='-', color='orange', label='Tempo 26 processos Dynamic')
plt.plot(threads, tempos_26g, marker='o', linestyle='-', color='purple', label='Tempo 26 processos Guided')
# Configurar os rótulos e título
plt.xlabel('Número de Threads')
plt.ylabel('Tempo (segundos)')
plt.title('Gráfico de Linhas: threads vs Tempo')
plt.grid(True)
plt.legend()

# Exibir o gráfico
plt.show()
```

```python
threads = [1, 4, 8]  # Número de threads
speedup_MPI = [temposMPI[0] / t for t in temposMPI]

speedup_1s = [tempos_1s[0] / t for t in tempos_1s]
speedup_1g = [tempos_1g[0] / t for t in tempos_1g]
speedup_1d = [tempos_1d[0] / t for t in tempos_1d]
speedup_13s = [tempos_13s[0] / t for t in tempos_13s]
speedup_13g = [tempos_13g[0] / t for t in tempos_13g]
speedup_13d = [tempos_13d[0] / t for t in tempos_13d]
speedup_26s = [tempos_26s[0] / t for t in tempos_26s]
speedup_26g = [tempos_26g[0] / t for t in tempos_26g]
speedup_26d = [tempos_26d[0] / t for t in tempos_26d]
# Criar o gráfico de linhas
plt.figure(figsize=(8, 5))
plt.plot(threads[:3], speedup_1s, marker='o', linestyle='-', color='b', label='Speedup 1 processo Static')
plt.plot(threads[:3], speedup_1d, marker='o', linestyle='-', color='r', label='Speedup 1 processo Dynamic')
plt.plot(threads[:3], speedup_1g, marker='o', linestyle='-', color='g', label='Speedup 1 processo Guided')
plt.plot(threads[:3], speedup_13s, marker='o', linestyle='-', color='c', label='Speedup 13 processos Static')
plt.plot(threads[:3], speedup_13d, marker='o', linestyle='-', color='m', label='Speedup 13 processos Dynamic')
plt.plot(threads[:3], speedup_13g, marker='o', linestyle='-', color='y', label='Speedup 13 processos Guided')
plt.plot(threads[:3], speedup_26s, marker='o', linestyle='-', color='k', label='Speedup 26 processos Static')
plt.plot(threads[:3], speedup_26d, marker='o', linestyle='-', color='orange', label='Speedup 26 processos Dynamic')
plt.plot(threads[:3], speedup_26g, marker='o', linestyle='-', color='purple', label='Speedup 26 processos Guided')
# plt.plot(threads_mpi, speedup_MPI, marker='o', linestyle='-', color='blue', label='Speedup MPI puro')
# Configurar os rótulos e título
plt.xlabel('Número de Threads')
plt.ylabel('Fator de Speedup')
plt.title('Gráfico de Linhas: Threads vs Speedup')
plt.grid(True)
plt.legend()

# Exibir o gráfico
plt.show()
```

```python
# Calcular eficiência para o MPI puro
eficiencia_MPI = [(speedup / thread) * 100 for speedup, thread in zip(speedup_MPI, threads_mpi)]

# Criar o gráfico de eficiência MPI
plt.figure(figsize=(8, 5))
plt.plot(threads_mpi, eficiencia_MPI, marker='o', linestyle='-', color='b', label='Eficiência MPI puro')

plt.xlabel('Número de Processos')
plt.ylabel('Eficiência (%)')
plt.title('Eficiência MPI puro: Processos vs Eficiência')
plt.grid(True)
plt.legend()
plt.show()
```
